```python
from qiskit_finance.circuit.library import NormalDistribution
```

```python
center_key = "0111000000"

num_qubits = 10
mu = int(center_key, 2) 
sigma_var = 15**2
bounds = (0, 2**num_qubits - 1)

normal_dist = NormalDistribution(num_qubits, mu, sigma_var, bounds)
qc = normal_dist.decompose()
```

```python
mu
```

```python
qc.draw('mpl', fold=-1)
```

```python
from qiskit.quantum_info import Statevector

sv = Statevector.from_instruction(qc)
probs = sv.probabilities_dict()
```

```python
import matplotlib.pyplot as plt
plt.bar([int(key, 2) for key in probs.keys()], probs.values())
plt.xlabel('State')
plt.ylabel('Probability')
plt.title('Probability Distribution of Normal Distribution Circuit')
plt.xticks(rotation=90)
plt.show()
```
